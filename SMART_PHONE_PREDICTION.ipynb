```python
#Importing library files
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error, mean_absolute_error
import tkinter as tk
from tkinter import ttk
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df=pd.read_csv('Smartphones_mod-1.csv')
# Ref for dataset: https://www.kaggle.com/code/imrulanwar836/prepared-features 
price_ranges = [0, 20000,60000,150000] # Adjust the ranges as needed
price_labels = ['0', '1', '2']

# Create a new column 'price_cluster' based on the price ranges
df['price_range'] = pd.cut(df['price'], bins=price_ranges, labels=price_labels, right=False)
df=df.dropna()
df.describe()

# for c in df:
#     print(c)
nan_values_in_column = df['price_range'].isna().any()

print(nan_values_in_column)
```

```python
# Importing dataframe for model 1 that classifies the phone price
lab = ['rating','Brand','Unnamed: 0', 'price', 'Brand1', 'os1', 'ProcessorName1','Processor Name','os','model','price_range']
X = df.drop(columns=lab)  # Features for the model
print(X.columns)
Y= df[['price_range']]  # Target variable
print(Y.shape)
print(X.shape)
```

```python
# Split the data into training, validation, and test sets
X_train, X_interm, Y_train, Y_interm = train_test_split(X, Y, test_size=0.35, random_state=42)
X_test, X_val, Y_test, Y_val = train_test_split(X_interm, Y_interm, test_size=0.3, random_state=42)
print("X_train",X_train.shape)
print("Y_train",Y_train.shape)
print("X_test",X_test.shape)
print("Y_test",Y_test.shape)
print("X_val",X_val.shape)
print("Y_val",Y_val.shape)
```

```python
# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)
print("X_train_scaled:\n",X_train_scaled)
print("X_test_scaled :\n",X_test_scaled)
print("X_val_scaled :\n",X_val_scaled)
```

```python
print("Train:",type(X_train_scaled),X_train_scaled.shape)
print("Test:",type(X_test_scaled),X_test_scaled.shape)
print("Validation:",type(X_val_scaled),X_val_scaled.shape)
```

```python
# Convert Y_train to a 1-dimensional array
Y_train = np.ravel(Y_train)
```

```python
# Train the logistic regression model
model = LogisticRegression(multi_class='multinomial', solver='lbfgs')
model.fit(X_train_scaled, Y_train)
```

```python
# Predict on the validation set
Y_val_pred = model.predict(X_val_scaled)
```

```python
# Calculate the accuracy on the validation set
val_accuracy = accuracy_score(Y_val, Y_val_pred)
print(f"Validation Accuracy: {val_accuracy}")
```

```python
#Model for low price model classification
lab_1= ['rating','Brand','Unnamed: 0', 'price', 'Brand1', 'os1', 'ProcessorName1','Processor Name','os','model','price_range']
X_1= low.drop(columns=lab)  # Features for the model
print(X_1.columns)
Y_1= low[['price']]  # Target variable
print(X_1.shape)
print(Y_1.shape)
```

```python
# Split the data into training, validation, and test sets
X_train_1, X_interm_1, Y_train_1, Y_interm_1 = train_test_split(X_1, Y_1, test_size=0.35, random_state=42)
X_test_1, X_val_1, Y_test_1, Y_val_1 = train_test_split(X_interm_1, Y_interm_1, test_size=0.3, random_state=42)
print("X_train_1",X_train_1.shape)
print("Y_train_1",Y_train_1.shape)
print("X_test_1",X_test_1.shape)
print("Y_test_1",Y_test_1.shape)
print("X_val_1",X_val_1.shape)
print("Y_val_1",Y_val_1.shape)
```

```python
# Create the linear regression model
model_low = LinearRegression()
```

```python
# Fit the model on the training data
model_low.fit(X_train_1, Y_train_1)

# Make predictions on the training data
Y_train_pred_1 = model_low.predict(X_train_1)

# Make predictions on the validation data
Y_val_pred_1= model_low.predict(X_val_1)

# Make predictions on the test data
Y_test_pred_1= model_low.predict(X_test_1)

# Calculate the mean squared error on the training data
mse_train_1= mean_squared_error(Y_train_1, Y_train_pred_1)

# Calculate the mean squared error on the validation data
mse_val_1= mean_squared_error(Y_val_1, Y_val_pred_1)

# Calculate the mean squared error on the test data
mse_test_1= mean_squared_error(Y_test_1, Y_test_pred_1)

# Calculate the coefficient of determination (R^2) on the training data
r2_train_1= r2_score(Y_train_1, Y_train_pred_1)

# Calculate the coefficient of determination (R^2) on the validation data
r2_val_1= r2_score(Y_val_1, Y_val_pred_1)

# Calculate the coefficient of determination (R^2) on the test data
r2_test_1= r2_score(Y_test_1, Y_test_pred_1)
```

```python
# Print the metrics
print("Mean Squared Error (MSE) - Training: ", mse_train_1)
print("Mean Squared Error (MSE) - Validation: ", mse_val_1)
print("Mean Squared Error (MSE) - Test: ", mse_test_1)
print("Coefficient of Determination (R^2) - Training: ", r2_train_1)
print("Coefficient of Determination (R^2) - Validation: ", r2_val_1)
print("Coefficient of Determination (R^2) - Test: ", r2_test_1)
```

```python
import tkinter as tk
from tkinter import ttk
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

# Assuming you have already loaded the data into `X_train_scaled`, `y_train_1`, `scaler`, `model`, `new_data`
# and have divided the data into `X_train_1`, `y_train_1`, `X_train_2`, `y_train_2`, `X_train_3`, `y_train_3` for the categorized data.

def predict_price():
    # Step 1: Get user input and store it in new_data
    features = [float(entry.get()) for entry in entries]
    new_data = pd.DataFrame([features], columns=labels)

    # Step 2: Scale the features
    new_data_scaled = scaler.transform(new_data)

    # Step 3: Use the trained model to make predictions
    new_price_range_predictions = model.predict(new_data_scaled)
    print("New Price Range Predictions:")
    print(new_price_range_predictions)

    # Convert predictions to integers
    new_price_range_predictions = new_price_range_predictions.astype(int)

    # Create empty dataframes for each category
    category_0_data = pd.DataFrame(columns=new_data.columns)
    category_1_data = pd.DataFrame(columns=new_data.columns)
    category_2_data = pd.DataFrame(columns=new_data.columns)

    # Loop through each sample in the new data and assign it to the respective category dataframe
    for i, prediction in enumerate(new_price_range_predictions):
        if prediction == 0:
            category_0_data = pd.concat([category_0_data, new_data.iloc[[i]]])
        elif prediction == 1:
            category_1_data = pd.concat([category_1_data, new_data.iloc[[i]]])
        elif prediction == 2:
            category_2_data = pd.concat([category_2_data, new_data.iloc[[i]]])

    # Predict the categorized data using respective models
    # Create the linear regression model for each category
    model_low = LinearRegression()
    model_med = LinearRegression()
    model_high = LinearRegression()

    # Make predictions for each category if it is not empty
    if not category_0_data.empty:
        model_low.fit(X_train_1, Y_train_1)
        predicted_low_prices = model_low.predict(category_0_data)
    else:
        predicted_low_prices = [0]

    if not category_1_data.empty:
        model_med.fit(X_train_2, Y_train_2)
        predicted_med_prices = model_med.predict(category_1_data)
    else:
        predicted_med_prices = [0]

    if not category_2_data.empty:
        model_high.fit(X_train_3, Y_train_3)
        predicted_high_prices = model_high.predict(category_2_data)
    else:
        predicted_high_prices = [0]

    # Convert the elements of lists to Python floats
    predicted_low_price = float(predicted_low_prices[0])
    predicted_med_price = float(predicted_med_prices[0])
    predicted_high_price = float(predicted_high_prices[0])

    # Display the predicted prices
    if predicted_low_price > 0:
        result_label.config(text=f"Predicted Price Range: \nLow: ${predicted_low_price:.2f}")
    elif predicted_med_price > 0:
        result_label.config(text=f"Predicted Price Range: \nMedium: ${predicted_med_price:.2f}")
    elif predicted_high_price > 0:
        result_label.config(text=f"Predicted Price Range: \nHigh: ${predicted_high_price:.2f}")
    else:
        result_label.config(text="No prediction available for the given input.")

def login():
    # Replace 'your_username' and 'your_password' with the actual username and password
    username = "shgo"
    password = "1234"

    entered_username = username_entry.get()
    entered_password = password_entry.get()

    if entered_username == username and entered_password == password:
        login_window.destroy()
        root.deiconify()
    else:
        login_status_label.config(text="Invalid Username or Password")

# Create the login window
login_window = tk.Tk()
login_window.title("Login Form")
login_window.attributes("-fullscreen", True)  # This line makes the login window fullscreen

# Center the login window on the screen
screen_width = login_window.winfo_screenwidth()
screen_height = login_window.winfo_screenheight()
login_window.geometry(f"{screen_width}x{screen_height}+0+0")

# Set the background color of the login window
login_window.configure(bg="yellow")

# Create a style object for login widgets
login_style = ttk.Style(login_window)
login_style.configure('Login.TLabel', font=('Helvetica', 14, 'bold', 'underline'), foreground='blue')
login_style.configure('Login.TEntry', font=('Helvetica', 14),foreground='black')
login_style.configure('Login.TButton', font=('Helvetica', 14, 'bold'), foreground='black', background='orange')

# Create a frame to hold the login widgets
login_frame = ttk.Frame(login_window)
login_frame.pack(expand=True)

# Create the login heading with an underline
ttk.Label(login_frame, text="LOGIN PAGE", style='Login.TLabel').grid(row=0, column=0, columnspan=2, padx=5, pady=10)

# Create labels and entry widgets for username and password
ttk.Label(login_frame, text="Username:", style='Login.TLabel').grid(row=1, column=0, padx=5, pady=5)
username_entry = ttk.Entry(login_frame, style='Login.TEntry')
username_entry.grid(row=1, column=1, padx=5, pady=5)

ttk.Label(login_frame, text="Password:", style='Login.TLabel').grid(row=2, column=0, padx=5, pady=5)
password_entry = ttk.Entry(login_frame, show="*", style='Login.TEntry')
password_entry.grid(row=2, column=1, padx=5, pady=5)

# Create the Login button
login_button = ttk.Button(login_frame, text="Login", command=login, style='Login.TButton')
login_button.grid(row=3, column=0, columnspan=2, padx=5, pady=10)

# Create a label to display login status
login_status_label = ttk.Label(login_frame, text="", style='Login.TLabel')
login_status_label.grid(row=4, column=0, columnspan=2, padx=5, pady=5)

# Hide the main window until the user logs in
root = tk.Tk()
root.title("Prediction Page")
root.attributes("-fullscreen", True)  # This line makes the prediction window fullscreen

# Center the prediction window on the screen
root.geometry(f"{screen_width}x{screen_height}+0+0")

# Set the background color of the prediction window
root.configure(bg="yellow")  # Different background color

# Create a style object for prediction widgets
predict_style = ttk.Style(root)
predict_style.configure('Predict.TLabel', font=('Helvetica', 14, 'bold', 'underline'), foreground='black')  # Different color for the heading
predict_style.configure('Predict.TEntry', font=('Helvetica', 14))
predict_style.configure('Predict.TButton', font=('Helvetica', 14, 'bold'), foreground='black', background='green')  # Different color for the button

# Create a frame to hold the prediction widgets
predict_frame = ttk.Frame(root)
predict_frame.pack(expand=True)

# Create the prediction heading with an underline
ttk.Label(predict_frame, text="PREDICTION PAGE", style='Predict.TLabel').grid(row=0, column=0, columnspan=2, padx=5, pady=10)

# Create labels and entry widgets for each feature
labels = list(new_data.keys())  # Exclude 'Price' column
entries = []
for i, label in enumerate(labels):
    ttk.Label(predict_frame, text=label, style='Predict.TLabel').grid(row=i+1, column=0, padx=5, pady=5)
    entry = ttk.Entry(predict_frame, style='Predict.TEntry')
    entry.grid(row=i+1, column=1, padx=5, pady=5)
    entries.append(entry)

# Create the Predict button
predict_button = ttk.Button(predict_frame, text="Predict", command=predict_price, style='Predict.TButton')
predict_button.grid(row=len(labels)+1, column=0, columnspan=2, padx=5, pady=10)

# Create a label to display the predicted price range
result_label = ttk.Label(predict_frame, text="", style='Predict.TLabel')
result_label.grid(row=len(labels)+2, column=0, columnspan=2, padx=5, pady=5)

# Hide the prediction window until the prediction is done
root.withdraw()

# The root.mainloop() call will ensure the GUI remains interactive
root.mainloop()
```
